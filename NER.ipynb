```python
import spacy
```

```python
nlp = spacy.load(name='en_core_web_md')
```

```python
doc=nlp('Jonathan is looking for a job in Turkey for 50k. World Health Organization WHO is useless')
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_)
```

```python
from spacy import displacy
```

```python
displacy.render(doc, style='dep')
```

```python
text = '''When Sebastian Thrun started working on self-driving cars at Google in 2007, few people outside of the company took him seriously.
He immigrated in 2003. Zafer loves NLP'''
```

```python
doc = nlp(text)
```

```python
displacy.render(doc, style='ent')
```

# BERT GOOGLE - BART FACEBOOK (KELİMERLİ EMBADING YAPIYOR) - YAZILARI VEKTOR HALİNE GETİRİYOR, BAG OF WORDS

```python
text = "The company launched a new AI platform for enterprise customers."
```

```python
labels = ["technology", "sports", "finance"]
```

```python
from transformers import pipeline
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
text = "This tutorial explains how transformer models are used in NLP."
candidate_labels = ["technology", "health", "sports", "finance"]
result = classifier(text, candidate_labels)
print(f"Top prediction: {result['labels'][0]} ({result['scores'][0]:.2%})")
```
